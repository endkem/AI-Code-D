9. Can you explain the difference between deep and shallow copy in Python?
Ans. A shallow copy generates a new object but refers to the same elements as the original, but a deep copy generates a new object with its own copy of the elements. Deep copy generates a completely independent copy, whereas shallow copy generates a new object that shares some or all of its data with the original.

19. How do you sort a list in Python?
Ans. Using the sort() method or the sorted() function, you can sort a list in Python. You can sort the items on a list either in ascending order or in descending order using either method.
A.	Use of the sort() method
The sort() function is a built-in method that alters the original list by sorting it in-place.
By default, the list's elements are rearranged in ascending order.
List.sort() is the syntax for the sort() method.
Here's an example:
numbers = [4, 2, 1, 3]
numbers.sort()
print(numbers)  
# Output: [1, 2, 3, 4]
B.	Use of the sorted() function
The built-in function sorted() returns a new sorted list while keeping the original list unaltered.
The components in the new list are by default arranged in ascending order.
The sorted() function can be used by using the syntax sorted(list).
Here's an example:
numbers = [4, 2, 1, 3]
sorted_numbers = sorted(numbers)
print(sorted_numbers) 
 # Output: [1, 2, 3, 4]
